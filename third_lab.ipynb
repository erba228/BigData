```python
import numpy as np

x = np.linspace(-4*np.pi, 4*np.pi, 100)

xlab = 'время сек'
ylab = 'амплитуда сигнала'

y1 = np.sin(2*x)*np.exp(-0.1*x)
leg1 = 'exp(-0.1*t)*sin(2*t)'

y2 = 10*np.sin(0.5*x)+np.cos(10*x)
leg2 = 'смешаный сигнал'

y3 = np.sin(x)
leg3 = 'опорный сигнал'

plt.grid(color = 'b', linewidth = 1)

plt.xlabel(xlab, fontsize='x-large')
plt.ylabel(ylab, fontsize='x-large')

plt.title('Гармонические сигналы')

plt.plot(x,y1,'b-',label=leg1)
plt.plot(x,y2,'g:',label=leg2)
plt.plot(x,y3,'r--',label=leg3)

plt.legend(loc='best')

plt.grid(True)

plt.savefig('signa13.png', format='png')
plt.show()
```

```python
import matplotlib as mpl

# Значение по оси X время, t

X = np.linspace(0, 4*np.pi, 400)

# Значения по осям Y_01 & Y_02

leg_01 = 'exp(-0.1*t)*sin(2*t)'
Y_01 = np.sin(2*X)*np.exp(-0.1*X)

# "модулированный" сигнал

leg_02 = '10*sin(0.5*t)+cos(10*t)'
Y_02 = 10*np.sin(0.5*X)+np.cos(10*X)

# Зададим размеры изображения диаграммы

mpl.rcParams['figure.figsize'] = (8.0, 6.0)

# Строим диаграмму
# Получаем ссылку на объект типа matplotlib.axes.AxesSubplot, текущую диаграмму

ax_01 = plt.axes()
line_01, = ax_01.plot(X, Y_01, 'b-')

ax_01.axis([-1, 13, -1.1, 1.1])
ax_01.grid(color='b', linewidth=1)
ax_01.set_xlabel('Время t sec', fontsize='x-large')
ax_01.set_ylabel(leg_01, color='b', fontsize='x-large')
ax_01.set_title('Гармонические сигналы '+leg_01+" и "+leg_02)

ax_02 = ax_01.twinx()
line_02, = ax_02.plot(X, Y_02, 'r-')

ax_02.axis([-1, 13, -12, 12])
ax_02.set_ylabel(leg_02, color='r', fontsize='x-large')
ax_02.legend((line_01, line_02), (leg_01, line_02), loc='best')
ax_02.grid(color='r')

plt.show()
```

Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 12, 200)  

a = -2
b = 4
c = 7
y = a*x**2 + b*x + c

plt.figure(figsize=(10, 6))  

plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')

plt.xlabel('Значения по оси X')
plt.ylabel('Значения по оси Y')
plt.title('График квадратичной функции')

plt.legend()

plt.grid(True)  
plt.show()
```

Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
x = np.linspace(-5, 5, 200)

A = 5  
B = 5  
C = 0  
y = A * np.sin(B * x + C)

plt.figure(figsize=(10, 6))  

plt.plot(x, y, label=f'y = {A} * sin({B} * x + {C})')

plt.xlabel('Значения по оси X')
plt.ylabel('Значения по оси Y')
plt.title('График гармонической функции')

plt.legend()

plt.grid(True)  
plt.show()
```

Вывести графики в разные поля при помощи функции subplot()

```python
x1 = np.linspace(-10, 12, 200)
a = -2
b = 4
c = 7
y1 = a * x1**2 + b * x1 + c

x2 = np.linspace(-5, 5, 200)
A = 5 
B = 5  
C = 0  
y2 = A * np.sin(B * x2 + C)

plt.figure(figsize=(8, 6))

# Первый график
plt.subplot(1, 2, 1)  
plt.plot(x1, y1, label=f'y = {a}x^2 + {b}x + {c}')
plt.xlabel('Значения по оси X')
plt.ylabel('Значения по оси Y')
plt.title('График квадратичной функции')
plt.legend()
plt.grid(True)

# Второй график
plt.subplot(1, 2, 2)  
plt.plot(x2, y2, label=f'y = {A} * sin({B} * x + {C})')
plt.xlabel('Значения по оси X')
plt.ylabel('Значения по оси Y')
plt.title('График гармонической функции')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Построить графики в одной координатной плоскости, сделать подписи графиков, указать легенду

```python
x1 = np.linspace(-10, 12, 200)
a = -2
b = 4
c = 7
y1 = a * x1**2 + b * x1 + c

x2 = np.linspace(-10, 12, 200)
A = 5  
B = 5  
C = 0 
y2 = A * np.sin(B * x2 + C)

plt.figure(figsize=(8, 6))

# Первый график
plt.plot(
    x1, y1, 
    label=f'y = {a}x^2 + {b}x + {c}', color='blue'
)

# Второй график
plt.plot(
    x2, y2, 
    label=f'y = {A} * sin({B} * x + {C})', color='red'
)

plt.xlabel('Значения по оси X')
plt.ylabel('Значения по оси Y')
plt.title('График 2-х функций')


plt.legend()
plt.grid(True)
plt.show()
```

```python
years = [x for x in range(2012,2023)]
vals = [
    4341.15, 4599.21, 4981.51, 5182.99, 
    4794.34, 4900.79, 4792.54, 4806.32, 
    5358.53, 6268.31, 7178.32
]

plt.figure(figsize=(10, 6))

bars = plt.bar(years, vals)

for bar, val in zip(bars, vals):
    plt.text(
        bar.get_x() + bar.get_width() / 2 - 0.15, val + 100, 
        str(round(val, 2)), 
        fontsize=10, ha='center', va='bottom', color='black'
    )

plt.xlabel('Год')
plt.ylabel('Сумма прожиточного минимума')
plt.title('График прожиточного минимума на период (2012-2022)')

plt.xticks(rotation=45)
plt.ylim(min(vals) - 200, max(vals) + 200)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

Построить круговой график по предыдущим данным

```python
years = [x for x in range(2012,2023)]
vals = [
    4341.15, 4599.21, 4981.51, 5182.99, 
    4794.34, 4900.79, 4792.54, 4806.32, 
    5358.53, 6268.31, 7178.32
]

dpi = 100

fig = plt.figure(
    dpi=dpi, 
    figsize=(640 / dpi, 480 / dpi)
)

mpl.rcParams.update({'font.size': 12})
plt.title('График прожиточного минимума на период (2012-2022)')


xs = range(len(years))

plt.pie(
    vals, autopct='%.2f', radius=1.2, 
    startangle=140, pctdistance=0.85,  
    explode=[0.15] + [0 for _ in range(len(years) - 1)]
)

plt.legend(
    bbox_to_anchor=(1, 0.5),
    loc='center left', labels=years
)

plt.show()
```

Построить график произвольной каркасной поверхности.


```python
years = [x for x in range(2012,2023)]
vals = [
    4341.15, 4599.21, 4981.51, 5182.99, 
    4794.34, 4900.79, 4792.54, 4806.32, 
    5358.53, 6268.31, 7178.32
]

fig = plt.figure(
    figsize=(10, 8)
)
ax = fig.add_subplot(111, projection='3d')

x = years
y = np.arange(0, len(years), 1)
x, y = np.meshgrid(x, y)
z = np.zeros_like(x)
for i in range(len(years)):
    z[:, i] = vals[i]

surf = ax.plot_surface(
    x, y, z, 
    cmap='inferno', 
    linewidth=0, 
    antialiased=False
)

ax.set_xlabel('Год')
ax.set_ylabel('Индекс')
ax.set_zlabel('Значение')
ax.set_title('График прожиточного минимума на период (2012-2022)')

plt.show()
```
